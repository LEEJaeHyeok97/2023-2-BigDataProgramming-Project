```python
import pandas as pd
import numpy as np
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
import re
from bs4 import BeautifulSoup

driver = webdriver.Chrome()
driver.get("https://www.melon.com/chart/age/index.htm?chartType=YE&chartGenre=KPOP&chartDate=2021")
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# 년도별 노래 데이터 최종 결과 리스트
dfs = []

def melon_collector(url, year):
    time.sleep(5)
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    # 제목 가져오기
    title=driver.find_elements(by=By.CLASS_NAME, value='ellipsis.rank01') # 제목

    title2=[]
    for i in title:
        title2.append(i.text)

    del title2[10:]    
    
    # 가수 가져오기
    singer=driver.find_elements(by=By.CLASS_NAME, value='ellipsis.rank02') # 가수

    singer2=[]
    for i in singer:
        singer2.append(i.text)

    del singer2[10:]
    
    # 가사, 장르 가져오기
    
    song_info = soup.find_all('div', attrs={'class': 'ellipsis rank01'})
    
    # Top 10만 추출
    songid = []

    for i in range(10):
        try:
            songid.append(re.sub(r'[^0-9]', '', song_info[i].find("a")["href"][43:])) # 곡 id정보 추출
        except:
            songid.append('')
            continue # 1992년 먼지가 되어와 같이 멜론에서 현재 들을 수 없는 노래 제외
        
        
    lyrics=[]
    genres=[]
    
    for i in songid:
        try:
            
            # songid가 없을 경우 lyrics와 개수 맞추기 위해 genres 추가
            genres_append = False
            if(not i):
                genres.append('')
                genres_append = True
                
            driver.get("https://www.melon.com/song/detail.htm?songId=" + i)

            html_content = driver.page_source
            # BeautifulSoup을 사용하여 HTML 파싱
            soup = BeautifulSoup(html_content, 'html.parser')
            
            time.sleep(2)
            
            if(genres_append == False):
                genre_element = soup.find('dt', text='장르')
                genre = genre_element.find_next('dd').text.strip()
                genres.append(genre)
            
            lyric=driver.find_element(by=By.CLASS_NAME, value="lyric")
            lyrics.append(lyric.text)
        except: # 가사가 없는경우
            lyrics.append('')
            continue
    lyrics2 = []
    years = []
    
    for i in lyrics:
        lyrics2.append(i.replace("\n"," "))
        years.append(year) 
    
    df=pd.DataFrame({"제목":title2, "가수":singer2, "장르":genres, "연도":years, "가사":lyrics2})
    dfs.append(df)
```

```python
start = 1964
url = 'https://www.melon.com/chart/age/index.htm?chartType=YE&chartGenre=KPOP&chartDate='

while start<=2022:
    new_url = url + str(start)
    melon_collector(new_url, start)
    print(start,'완료') # 완료 여부 표시
    start += 1
    
final_df = pd.concat(dfs, ignore_index=True)
final_df.to_csv('멜론_전체.csv', index=False, encoding='utf-8-sig')
```
